```python
import pathlib
data_dir = pathlib.Path(data_dir)
type(data_dir)
```

```python
print(list(data_dir.glob("*")))
print(len(list(data_dir.glob("*"))))
print(len(list(data_dir.glob("Dog/*"))))
print(len(list(data_dir.glob("Cat/*"))))
print(len(list(data_dir.glob("*/*"))))
```

```python
import PIL
num_skipped = 0
for folder_name in ("Cat", "Dog"):
    folder_path = os.path.join(data_dir, folder_name)
    for fname in os.listdir(folder_path):
        fpath = os.path.join(folder_path, fname)
        try:
            with open(fpath, "rb") as fobj:
                is_jfif = tf.compat.as_bytes("JFIF") in fobj.peek(10)
            if not is_jfif:
                raise ValueError("Not a valid image file")
        except:
            num_skipped += 1
            os.remove(fpath)
print(f"Deleted {num_skipped} corrupted images.")
```

```python
cat_dog_images_dict = {
    'Cat': list(data_dir.glob('Cat/*')),
    'Dog': list(data_dir.glob('Dog/*')),

}
cat_dog_labels_dict = {
    'Cat': 0,
    'Dog': 1,
}
```

```python
X, y = [], []

for animal_name, images in cat_dog_images_dict.items():
    for image in images:
        img = cv2.imread(str(image))
        if img is not None:
            resized_img = cv2.resize(img, (224, 224))
            X.append(resized_img)
            y.append(cat_dog_labels_dict[animal_name])
        else:
            print(f"Could not read image: {image}")  
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2  
)

train_gen = datagen.flow_from_directory(
    'PetImages',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='training'
)

val_gen = datagen.flow_from_directory(
    'PetImages',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)
```

```python
import tensorflow as tf
import tensorflow_hub as hub

import tensorflow_hub as hub
m = tf.keras.Sequential([
    Lambda(lambda x:hub.KerasLayer("https://www.kaggle.com/models/google/mobilenet-v2/TensorFlow2/035-224-feature-vector/2",
                   trainable=False)(x)),  
    tf.keras.layers.Dense(1, activation='sigmoid')
])
m.build([None, 224, 224, 3])
```

```python
m.summary()
```

```python
m.compile(
    optimizer="adam",
    loss=tf.keras.losses.BinaryCrossentropy(),  
    metrics=['accuracy']  
)
```

```python
history = m.fit(
    train_gen,
    validation_data=val_gen,
    epochs=5
)
```

```python
from tensorflow.keras.preprocessing import image
img = image.load_img('dog 3.jpg', target_size=(224, 224))
img_array = image.img_to_array(img) / 255.0
img_array = np.expand_dims(img_array, axis=0)

predictions = m.predict(img_array)

if predictions[0][0] > 0.5:
    print("The new image is: Dog")
else:
    print("The new image is: Cat")
```
